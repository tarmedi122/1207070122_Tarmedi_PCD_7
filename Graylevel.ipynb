```python
import cv2 # untuk pemrosesan gambar, numpy untuk operasi matriks, scikit-image untuk membaca gambar, dan matplotlib.pyplot untuk visualisasi.
import numpy as np # untuk pemrosesan gambar, numpy untuk operasi matriks, scikit-image untuk membaca gambar, dan matplotlib.pyplot untuk visualisasi.
from skimage import io # untuk pemrosesan gambar, numpy untuk operasi matriks, scikit-image untuk membaca gambar, dan matplotlib.pyplot untuk visualisasi.
import matplotlib.pyplot as plt # untuk pemrosesan gambar, numpy untuk operasi matriks, scikit-image untuk membaca gambar, dan matplotlib.pyplot untuk visualisasi.

img = io.imread('Mobil 1.jpg') # Untuk membaca gambar dengan nama 'Mobil 1.jpg' menggunakan fungsi io.imread dari scikit-image dan menyimpannya ke dalam variabel img.

row, column, _ = img.shape # atribut shape dari array gambar. Variabel row menyimpan jumlah baris gambar, column menyimpan jumlah kolom, dan _ digunakan untuk mengabaikan dimensi kedalaman warna yang tidak diperlukan.

img1 = np.zeros((row, column), dtype='uint8') # membuat array numpy dengan ukuran yang sama dengan gambar awal, berisi nol, dan tipe data uint8 (unsigned 8-bit integer). Array ini akan digunakan untuk menyimpan citra hasil thresholding.

min_range = 10 # endefinisikan rentang nilai piksel yang akan digunakan untuk proses thresholding. Dalam hal ini, nilai piksel saluran merah harus berada di antara 10 dan 60 untuk dianggap memenuhi kriteria.
max_range = 60 # endefinisikan rentang nilai piksel yang akan digunakan untuk proses thresholding. Dalam hal ini, nilai piksel saluran merah harus berada di antara 10 dan 60 untuk dianggap memenuhi kriteria.

for i in range(row): #  iterasi pada setiap piksel gambar dan memeriksa nilai piksel pada saluran merah (indeks 0)
    for j in range(column): #  iterasi pada setiap piksel gambar dan memeriksa nilai piksel pada saluran merah (indeks 0)
        if img[i, j, 0] > min_range and img[i, j, 0] < max_range: # Jika nilai piksel berada di antara min_range dan max_range, piksel pada citra hasil (img1) diatur menjadi 255 (putih), jika tidak, piksel diatur menjadi 0 (hitam).
            img1[i, j] = 255 # Jika nilai piksel berada di antara min_range dan max_range, piksel pada citra hasil (img1) diatur menjadi 255 (putih), jika tidak, piksel diatur menjadi 0 (hitam).
        else: # Jika nilai piksel berada di antara min_range dan max_range, piksel pada citra hasil (img1) diatur menjadi 255 (putih), jika tidak, piksel diatur menjadi 0 (hitam).
            img1[i, j] = 0 # Jika nilai piksel berada di antara min_range dan max_range, piksel pada citra hasil (img1) diatur menjadi 255 (putih), jika tidak, piksel diatur menjadi 0 (hitam).
# Pada kodingan di bawah semua ini yaitu untuk mengatur tata letak subplot dan menampilkan gambar asli, histogramnya, citra hasil thresholding, dan histogramnya menggunakan matplotlib.pyplot. Subplot pertama menampilkan gambar asli dan histogramnya, sedangkan subplot kedua menampilkan citra hasil thresholding dan histogramnya. 
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()
# ini juga kebawahnya Menampilkan gambar asli pada subplot pertama, mengatur judulnya, dan menggambar histogram gambar pada subplot kedua dengan 256 bins.
ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img[:, :, 0].ravel(), bins=256)
ax[1].set_title('Histogram Input')
# ini juga kebawahnya Menampilkan gambar asli pada subplot pertama, mengatur judulnya, dan menggambar histogram gambar pada subplot kedua dengan 256 bins.
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')

plt.tight_layout() # # untuk menampilkan Gambar biar rapi
plt.show() # Untuk menampilkan suatu gambar
```
